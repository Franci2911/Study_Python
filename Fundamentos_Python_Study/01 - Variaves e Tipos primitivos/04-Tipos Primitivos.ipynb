Tipos **primitivos**

```python
n1 = int(input('Digite um numero: '))
n2 = int(input('Digite o segundo numero: '))
soma = n1 + n2
print(f'A soma dos dois numeros é: {soma}')
```

```python
v1 = input('Digite algo: ')
print(f'O tipo da variavel digitada é: {type(v1)}')
print(f'è uma letra ou um numero ? {v1.isalnum()}')
print(f'è uma letra ? {v1.isalpha()}')
print(f'é um numero com casa decimal ? {v1.isdecimal()}')
print(f'è um numero ? {v1.isnumeric()}')
print(f'todas as letras são minusculas ? {v1.islower()}')
print(f'tem algum espaço entra os caracteres? {v1.isspace()}')
print(f'è maiuscula ? {v1.isupper()}')
```
