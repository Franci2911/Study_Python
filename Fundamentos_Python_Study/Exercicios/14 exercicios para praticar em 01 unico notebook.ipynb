```python
# Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

ganhoHora1 =  float(input('quanto você ganha por hora: '))
horaTrabalhada1 = float(input('quantas horas você trabalha no mês: '))

#calcule o salario bruto:
salarioBruto = horasTrabalhada * ganhoHora1
print(f'O salario bruto é: {salarioBruto}')

# calcule o desconto do imposto de renda - IR
impostoRenda = salarioBruto - (salarioBruto * 11/100)
print(f'O salario bruto com desconto de imposta é: {impostoRenda}')


# Calcule o desconto do INSS (8% do salário bruto)
impostoRenda = (salarioBruto * 11/100)
impostoINSS = (salarioBruto * 8/100)
impostoSindicato = (salarioBruto * 5/100)

salarioLiquido  = (salarioBruto - impostoRenda) - impostoINSS - impostoSindicato 

print(f'O salario bruto com desconto de imposta é: {salarioLiquido}')
```

#### 18. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
tamanhoArquivo = float(input('digite o tamanho do arquivo: '))
velocidadeNET = float(input('digite a velocidade da sua internet em mbps: '))

# aqui trasnformamos megabytes em megabits para que tamanho e velocidade estejam na mesma unidade
tamanho_megabits = tamanhoArquivo * 8

tempo = tamanho_megabits / velocidadeNET
# transformando em minutos
tempo_minutos = tempo / 60
print(f'O tempo de download é de {tempo_minutos} minutos')

```
